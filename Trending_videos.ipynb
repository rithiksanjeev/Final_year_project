```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
sns.set_style('white') 
%matplotlib inline 
trend1 = pd.read_csv("/home/user/Trending.csv")
trend1.head
trend2 = pd.read_csv("/home/user/Trending2.csv")
trend2.head
a=pd.merge(trend1,trend2, on='@videoid')
a.sort_values(by = ['@viewcount','@likes'], ascending = False)
a.describe()
a['@videoid'].value_counts()
columnWeCareAbout=['@videoid']
a[columnWeCareAbout]
uniform_data = np.random.rand(10, 12)
ax = sns.heatmap(uniform_data)
sns.heatmap(a.isnull(), yticklabels=False, cbar=False, cmap='viridis')#used to check missing data
ratings = a.sort_values(by = ['@viewcount','@likes'], ascending = False) 
plt.figure(figsize =(10, 4))   
ratings['@viewcount'].hist(bins = 70)
plt.figure(figsize =(10,4))  
plt.xticks(rotation='vertical')
ratings['@videoid'].hist(bins = 100) 
trending_personal=ratings['@videoid'].unique()#if repaeted due to technical faults
print ("List of trending personlised")
trending_personal
pd.DataFrame(trending_personal).to_csv("/home/user/List_of_videos1.csv")
```
